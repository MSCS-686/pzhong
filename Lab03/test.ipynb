```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

X, Y = make_classification(200, 2, 2, 0, weights=[.5, .5], random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y)
```

```python
from perceptron import perceptron
```

```python
p = perceptron(learning_rate=0.01, epochs=10)
p.fit(X, Y)
```

```python
weights = p.get_weights()
bias = p.get_bias()
```

```python
x = np.linspace(-0.6, 1)
y = (-(bias/weights[1])/(bias/weights[0]) * x + (-bias/weights[1]))
plt.scatter(X[:,0], X[:,1], c=p.predict(X))
plt.plot(x, y, '--r')
```
