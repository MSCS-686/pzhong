```python
from sklearn.datasets.samples_generator import make_blobs
from kmeans import KMeansCluster
from matplotlib import pyplot as plt
```

```python
X, cluster_assignments = make_blobs(n_samples=200, centers=4, cluster_std=0.60, random_state=0)
```

```python
km = KMeansCluster(k=4)
cls, cen = km.fit(X)
```

```python
colors=['orange', 'blue', 'green', 'red']
for i in range(X.shape[0]):
    plt.scatter(X[i, 0], X[i, 1], s=7, color = colors[cls[i]])
plt.scatter(cen[:,0], cen[:,1], marker='*', c='g', s=150)
```
